# concatenate()

```python
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df =pd.concat([india_weather,us_weather])
df
```

```python
df =pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
df =pd.concat([india_weather,us_weather],keys=['India','US'])
df
```

```python
df.loc['India']
```

## concatenating using index

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
```

```python
df = pd.concat([temperature_df,windspeed_df], axis=1)
df
```

# merge()

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df = pd.merge(df1,df2,on='city')
df
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df = pd.merge(df1,df2,on='city',how='left') # same you can use inner, outer, right
df
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df = pd.merge(df1,df2,on='city',how='outer', indicator=True) # same you can use inner, outer, right
df
```

```python
df1.merge(df2,on='city',suffixes=('_left','_right'))
```
