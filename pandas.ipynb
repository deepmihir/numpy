# Habdle missing data : fillna, dropna, interpolate

```python
df = pd.read_csv("weather_data2.csv")
type(df.day[0])
```

```python
df = pd.read_csv("weather_data2.csv", parse_dates=['day'])
type(df.day[0])
df.set_index(['day'], inplace=True)
```

## fillna()

```python
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df = df.fillna(method="bfill") 
new_df
```

now I want to fil row vise using ffill and bfill

```python
new_df = df.fillna(method="ffill", axis=1)
new_df
```

also we can specify explicitly

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event':"no-event"
})
new_df
```

## interpolate()

```python
new_df = df.interpolate() #by default linear interpolation
new_df
```

```python
new_df = df.interpolate(method="time")
new_df
```

## dropna()

```python
new_df = df.dropna() #this simply drop all with contain na value
new_df
```

```python
new_df = df.dropna(how="all") # this will only drop rows with all na value
new_df
```

```python
new_df = df.dropna(thresh=1) #this simply drop atleast one contain na value
new_df
```

## filling missing date

```python
dt = pd.date_range('01-01-2017','01-11-2017')
idx = pd.DatetimeIndex(dt)

new_df = new_df.reindex(idx)
new_df
```
