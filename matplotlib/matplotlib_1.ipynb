#  Axes labels, Legend, Grid

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
plt.xlabel('Days')
plt.ylabel('Temp')
plt.title('Weather')
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend(loc='best',bbox_to_anchor=(0.5, 0., 0.5, 0.5),fancybox=True)
plt.grid(
)
```

# Barchart

```python
import numpy as np
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

xpos =  np.arange(len(company))
plt.ylabel('revenue')
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue,label='revenue',width=0.4)
plt.bar(xpos+0.2,profit,label='profit',width=0.4)
plt.legend()
```

```python
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```

# Histograms

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.xlabel('sugar level')
plt.ylabel('No. of patients')
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.5,color='g')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel('sugar level')
plt.ylabel('No. of patients')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','pink'],label=['Men','Women'])
plt.legend()
```

```python
plt.xlabel('sugar level')
plt.ylabel('No. of patients')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,histtype='step',orientation='horizontal')
plt.legend()
```

# Pie chart
